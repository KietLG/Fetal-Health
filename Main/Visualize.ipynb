```python
sns.countplot(x='fetal_health', data=df)
plt.title("fetal_health")
plt.show()
```

```python
corrmat= df.corr()
plt.figure(figsize=(15,15))  
cmap = sns.diverging_palette(250, 10, s=80, l=55, n=9, as_cmap=True)
sns.heatmap(corrmat,annot=True, cmap=cmap, center=0)
```
